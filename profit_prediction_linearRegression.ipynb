#OnehotEncoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh=OneHotEncoder(sparse=False,
                 handle_unknown="error",
                 drop='first')
dfn=oh.fit_transform(df[["Area"]])
```

```python
dffn=pd.DataFrame(dfn)
df=df.join(dffn)
df
```

```python
df=df.drop("Area",axis=1)
df
```

```python
x=df.drop("Profit",axis=1)
x
```

```python
y=df["Profit"]
```

#Train data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.2)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

#Model Generate

```python
from sklearn.linear_model import LinearRegression
```

```python
lir=LinearRegression()

lir.fit(xtrain,ytrain)
```

```python
print(f'Trainning Accuracy is {(lir.score(xtrain,ytrain)*100)} %')
print(f'Test Accuracy is {(lir.score(xtest,ytest)*100)} %')
```
